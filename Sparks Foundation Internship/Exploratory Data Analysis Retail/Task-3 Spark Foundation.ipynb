```python
plt.figure(figsize=(10,10))
for i in range(len(cat_cols)):
    plt.subplot(3,3,i+1)
    value=df[cat_cols[i]].value_counts()
    ax=value.plot(kind='bar')
    ax.bar_label(ax.containers[0])
    plt.xlabel(cat_cols[i])
    plt.ylabel('count')
    plt.title('bar Graph')
    plt.tight_layout()
```

```python
plt.figure(figsize=(15,10))
for i in range(len(cat_cols)):
    plt.subplot(4,3,i+1)
    values=df[cat_cols[i]].value_counts().values
    names=df[cat_cols[i]].value_counts().keys()
    plt.pie(x=values,labels=names,autopct="%0.2f%%",shadow=True,startangle=30)
    plt.xlabel(cat_cols[i])
    plt.title('pie chart')
    plt.tight_layout()
```

```python
q1=np.percentile(df['Sales'],25)
q2=np.percentile(df['Sales'],50)
q3=np.percentile(df['Sales'],75)
iqr=q3-q1
ub=q3+1.5*iqr
lb=q1-1.5*iqr
cond1=df['Sales']>ub
cond2=df['Sales']<ub
df['Sales'][cond1|cond2].values
len(df['Sales'][cond1|cond2].values)
```

```python
for i in range(len(num_cols)):
    q1 = np.percentile(df[num_cols[i]],25)
    q2 = np.percentile(df[num_cols[i]],50)
    q3 = np.percentile(df[num_cols[i]],75)
    iqr= q3-q1
    ub= q3+1.5*iqr
    lb= q1-1.5*iqr
    con1=df[num_cols[i]]>ub
    con2=df[num_cols[i]]<lb
    outliers=df[num_cols[i]][con1|con2].values
    print(num_cols[i],len(outliers))
```

```python
q1 = np.percentile(df['Sales'],25)
q2 = np.percentile(df['Sales'],50)
q3 = np.percentile(df['Sales'],75)
iqr= q3-q1
ub= q3+1.5*iqr
lb= q1-1.5*iqr
con1=df['Sales']>ub
con2=df['Sales']<lb
median=df['Sales'].median()
df['Sales']=np.where(con1|con2,median,df['Sales'])
```

```python
q1 = np.percentile(df['Quantity'],25)
q2 = np.percentile(df['Quantity'],50)
q3 = np.percentile(df['Quantity'],75)
iqr= q3-q1
ub= q3+1.5*iqr
lb= q1-1.5*iqr
con1=df['Quantity']>ub
con2=df['Quantity']<lb
median=df['Quantity'].median()
df['Quantity']=np.where(con1|con2,median,df['Quantity'])
```

```python
q1 = np.percentile(df['Profit'],25)
q2 = np.percentile(df['Profit'],50)
q3 = np.percentile(df['Profit'],75)
iqr= q3-q1
ub= q3+1.5*iqr
lb= q1-1.5*iqr
con1=df['Profit']>ub
con2=df['Profit']<lb
median=df['Profit'].median()
df['Profit']=np.where(con1|con2,median,df['Profit'])
```

```python
q1 = np.percentile(df['Discount'],25)
q2 = np.percentile(df['Discount'],50)
q3 = np.percentile(df['Discount'],75)
iqr= q3-q1
ub= q3+1.5*iqr
lb= q1-1.5*iqr
con1=df['Discount']>ub
con2=df['Discount']<lb
median=df['Discount'].median()
df['Discount']=np.where(con1|con2,median,df['Discount'])
```

```python
plt.figure(figsize=(10,3))
plt.suptitle('After removing of outliers')
for i in range(len(num_cols)):
    plt.subplot(2,3,i+1)
    plt.boxplot(df[num_cols[i]],vert=False)
    plt.title(num_cols[i])
    plt.tight_layout()
```

**Convert Categorical data into numerical**

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in cat_cols:
    df[i]=le.fit_transform(df[i])
```

```python
df.head()
```

**Standardize the data**

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
for i in df.columns[:-1]:
    df[i]=ss.fit_transform(df[[i]])
```
